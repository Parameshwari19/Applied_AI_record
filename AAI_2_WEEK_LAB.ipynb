```python
import numpy as np
```

```python
def step_function(x):
    return 1 if x >= 0 else 0
```

```python
def train_perceptron(X, T, lr, gate_name):
    w = np.zeros(X.shape[1])
    b = 0
    epoch = 0

    while True:
        epoch += 1
        total_error = 0

        for i in range(len(X)):
            yin = np.dot(X[i], w) + b
            y = step_function(yin)
            error = T[i] - y

            total_error += abs(error)

            # Update rule
            w = w + lr * error * X[i]
            b = b + lr * error

        if total_error == 0:
            print(f"{gate_name} Gate converged in {epoch} epochs")
            break

    return w, b
```

```python
n = int(input("Enter number of inputs (2 for AND/OR): "))
p = int(input("Enter number of input patterns: "))

X = []
print("\nEnter input patterns (space separated):")
for i in range(p):
    X.append(list(map(int, input(f"Pattern {i+1}: ").split())))

X = np.array(X)

print("\nEnter AND gate targets (space separated):")
T_AND = np.array(list(map(int, input().split())))

print("Enter OR gate targets (space separated):")
T_OR = np.array(list(map(int, input().split())))

lr = float(input("\nEnter learning rate: "))

w_and, b_and = train_perceptron(X, T_AND, lr, "AND")
print("Final AND Weights:", w_and)
print("Final AND Bias:", b_and)

w_or, b_or = train_perceptron(X, T_OR, lr, "OR")
print("Final OR Weights:", w_or)
print("Final OR Bias:", b_or)
```
